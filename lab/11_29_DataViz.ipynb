## Seaborn for plotting

Seaborn provides more functionality and organization than Matplotlib. Seaborn contains a number of patterns and plots for data visualization.
Behind the scenes, seaborn uses matplotlib, along with Numpy and Pandas to draw its plots

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips") #load dataset
```

```python
tips.head()
```
