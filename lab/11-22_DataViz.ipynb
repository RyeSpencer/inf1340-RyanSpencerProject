```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(1, 50)
Y = 3*X

# Plot lines and/or markers to the Axes.
plt.plot(X, Y)

# Display the figure.
plt.show()
```

```python
# customise the line graphs

plt.plot(X, Y)

# Set the x axis label of the current axis.
plt.xlabel('x - axis')

# Set the y axis label of the current axis.
plt.ylabel('y - axis')

# Set a title 
plt.title('Draw a line')

# Display the figure.
plt.show()
```

```python
plt.plot(X, Y, color='red', linewidth = 5, linestyle='--')

plt.title('Draw a line.')

plt.show()
```

### **Python program to display a bar chart of the popularity of programming Languages.**

Programming languages: Java, Python, PHP, JavaScript, C#, C++

Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(x, popularity, color='blue')
# plt.barh(x, popularity, color='blue') # horizontal bar chart
# plt.barh(x, popularity, color='0000') # hex code color parameter
# plt.bar(x, popularity, color=['red', 'black', 'green', 'blue', 'yellow', 'cyan']) # try color variations


plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")

plt.show()
```

### 8- Plot x in terms of y and y in terms of x in two different subplots of one figure.
x =[0, 1, 2, 3, 4, 5, 6]

y =[0, 1, 3, 6, 9, 12, 17]

```python
# defining the values of X 
x =[0, 1, 2, 3, 4, 5, 6] 
  
# defining the value of Y 
y =[0, 1, 3, 6, 9, 12, 17] 
  
# creating the canvas with class 'fig' 
# and it's object 'axes' with '1' row  
# and '2' columns 
fig, axes = plt.subplots(1, 2) 
  
axes[0].plot(x, y, 'r--o') # plots first graph
axes[1].plot(y, x, '-.') # plots second graph 
fig.tight_layout() 

#changing the layout to 2 rows and 1 column
fig, axes = plt.subplots(2, 1) # plots first graph
axes[0].plot(x, y, 'r--o') 
axes[1].plot(y, x, '-.') # plots second graph 
  
# Gives a clean look to the graphs 
fig.tight_layout() 
```

```python
#For 4 subplots

x =[0, 1, 2, 3, 4, 5, 6] 
y =[0, 1, 3, 6, 9, 12, 17]

fig, axes = plt.subplots(2, 2) 
  
axes[0,0].plot(x, y, 'g--o') 
axes[0,1].plot(y, x, '-.') 
axes[1,0].plot(y, x, 'm--o') 
axes[1,1].plot(y, x, '--') 
  
fig.tight_layout() 
```
