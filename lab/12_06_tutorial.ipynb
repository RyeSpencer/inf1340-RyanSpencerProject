```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
import pandas as pd
import plotly.express as px
```

https://datavizcatalogue.com/ is a useful webpage to understand different types of visualizations. 

```python
url = "https://raw.githubusercontent.com/shionguha/inf1340-programmingfordatascience-fa22/main/data/titanic_dataset.csv"
df = pd.read_csv(url)
df
```

```python
df.head()
```

```python
sns.set_style('white')

#we will be showing 
f, ax = plt.subplots(2, 2, figsize=(19,20))
f.suptitle('Plots showing..write in a title',fontweight='bold',fontsize=16)


ax[0,0].set_title('Line plots showing change in fare by age',fontweight="bold")
ax[0,0].set_ylabel('Age')
sns.lineplot(y="fare", x="age", hue = "sex", estimator = "mean", ci=None, data=df, ax=ax[0,0])

ax[0,1].set_title('Distribution of survival state depending on passenger class',fontweight="bold")
ax[0,1].set_ylabel('Survived')
sns.violinplot(y="survived", x="pclass", data=df, ax=ax[0,1])
ax[0,1].set_xticklabels(['1st', '2nd', '3rd'])

ax[1,0].set_title('Scatter plot showing the relationship between fare and age',fontweight="bold")
ax[1,0].set_ylabel('Fare')
sns.scatterplot(y="fare", x="age", data=df, ax=ax[1,0])

ax[1,1].set_title('Math second period',fontweight="bold")
ax[1,1].set_ylabel('Mean grade')
sns.countplot(x = df['survived'], ax=ax[1,1])

f.tight_layout()
f.subplots_adjust(top=.93)
plt.show()
```
