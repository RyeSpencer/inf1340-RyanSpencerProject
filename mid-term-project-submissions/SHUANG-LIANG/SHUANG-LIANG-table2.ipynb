```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
table2 = pd.read_excel(r'/Users/liangshuang/Desktop/table2.xlsx')

table2.head()
```

```python
table2 = table2.drop(labels = range(0, 15), axis=0)
table2
```

```python
table2 = table2.reset_index(drop = True)
table2
```

```python
table2.columns = ["Sort order", "Major area, region, country or area of destination", "Notes", "Country code", "1990B","1995B","2000B","2005B","2010B","2015B","1990M","1995M","2000M","2005M","2010M","2015M","1990F","1995F","2000F","2005F","2010F","2015F"]
table2
```

```python
# "Notes" in the third column is just symbols represent for comments and it should not include in the dataframe, we have an information sheet called "NOTES" that would explain those symbols
table2 = table2.drop(["Notes"], axis = 1)
table2
```

```python
#problem: there are multiple variables stored in 1 column
#tidy data principle #2: each column needs to consist of one and only one variable
table2.columns
```

```python
#problem: variables are stored in both rows and columns
#tidy data principle #3: variables need to be in cells, not rows and columns
table2.head()
```

```python
table2.tail()
```

```python
table2 = table2.melt(id_vars = ["Sort order", "Major area, region, country or area of destination", "Country code"], value_vars = ["1990B","1995B","2000B","2005B", "2010B","2015B","1990M","1995M", "2000M","2005M","2010M","2015M","1990F","1995F", "2000F","2005F","2010F","2015F"])
table2
```

```python
table2 = (table2.assign(Year = lambda x: x.variable.str[0:4].astype(int), Gender = lambda x: x.variable.str[4].astype(str)).drop("variable",axis = 1))
table2.head(10)
```

```python
# Styling the dataset
table2 = table2.replace({"Gender": {"B": "Both sexes", "F": "Female", "M": "Male"}})
table2.sample(10)
```

```python
table2 = table2.rename({"value": "Total population at mid-year (thousands)"}, axis = 1)
table2
```

```python
#change position of columns
table2 = table2.iloc[:, [0,1,2,4,5,3]]
table2
```

```python
table2.info()
```

```python
table2.tail()
```

```python
# There are some missing value but shows by "..", so we change to 0 that we can drop those in the next step
table2 = table2.replace("..", "NaN")

# The missing values are meaningless, so we drop those rows include missing values
table2 = table2.drop(table2[table2["Total population at mid-year (thousands)"] == "NaN"].index)

# Check whether the missing values deleted
table2.tail()
```
