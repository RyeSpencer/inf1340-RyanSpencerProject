```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
table5 = pd.read_excel(r'/Users/liangshuang/Desktop/table5.xlsx')

table5.head()
```

```python
table5 = table5.drop(labels=range(0, 15), axis=0)
table5
```

```python
table5 = table5.reset_index(drop = True)
table5
```

```python
table5.columns = ["Sort order", "Major area, region, country or area of destination", "Notes", "Country code", "Type of data (a)", "1990-1995B","1995-2000B","2000-2005B","2005-2010B","2010-2015B","1990-1995M","1995-2000M","2000-2005M","2005-2010M","2010-2015M","1990-1995F","1995-2000F","2000-2005F","2005-2010F","2010-2015F"]
table5
```

```python
# "Notes" in the third column is just symbols represent for comments and it should not include in the dataframe, we have an information sheet called "NOTES" that would explain those symbols
table5 = table5.drop(["Notes"], axis = 1)
table5
```

```python
#problem: there are multiple variables stored in 1 column
#tidy data principle #2: each column needs to consist of one and only one variable
table5.columns
```

```python
#problem: variables are stored in both rows and columns
#tidy data principle #3: variables need to be in cells, not rows and columns
table5.head()
```

```python
table5.tail()
```

```python
table5 = table5.melt(id_vars = ["Sort order", "Major area, region, country or area of destination", "Country code", "Type of data (a)"], value_vars = ["1990-1995B","1995-2000B","2000-2005B","2005-2010B","2010-2015B","1990-1995M","1995-2000M","2000-2005M","2005-2010M","2010-2015M","1990-1995F","1995-2000F","2000-2005F","2005-2010F","2010-2015F"])
table5
```

```python
table5 = (table5.assign(Year = lambda x: x.variable.str[0:9].astype(str), Gender = lambda x: x.variable.str[9].astype(str)).drop("variable", axis = 1))
table5.head(10)
```

```python
# Styling the dataset
table5 = table5.replace({"Gender": {"B": "Both sexes", "F": "Female", "M": "Male"}})
table5.sample(10)
```

```python
table5 = table5.rename({"value": "Annual rate of change of the migrant stock"}, axis = 1)
table5
```

```python
#change position of columns
table5 = table5.iloc[:, [0,1,2,3,5,6,4]]
table5
```

```python
table5.info()
```

```python
# There are some missing value but shows by "..", so we change to "NaN" that we can drop those in the next step
table5 = table5.replace("..", "NaN")

# The missing values are meaningless, so we drop those rows include missing values
table5 = table5.drop(table5[table5["Annual rate of change of the migrant stock"] == "NaN"].index)
```

```python
table5.dtypes
```

```python
table5 = table5.astype({"Annual rate of change of the migrant stock": "float"})
```

```python
table5.dtypes # The types are all correct now
```

```python
table5 # The values in the last column need to round to 2 decimal places
```

```python
# We can round the values now under the "Annual rate of change of the migrant stock" column since the types are all correct
table5["Annual rate of change of the migrant stock"] = table5["Annual rate of change of the migrant stock"].round(decimals = 2)
table5
```

```python
table5 = table5.reset_index(drop=True)
table5
```

```python
#problem: there are multiple types of data stored in 1 table
#tidy data principle #4: each table column needs to have a singular data type
table5.head()
```
