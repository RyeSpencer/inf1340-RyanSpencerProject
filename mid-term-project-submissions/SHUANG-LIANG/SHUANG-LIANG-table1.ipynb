```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
table1 = pd.read_excel(r'/Users/liangshuang/Desktop/table1.xlsx')

table1.head()
```

```python
table1 = table1.drop(labels = range(0, 15), axis=0)
table1
```

```python
table1 = table1.reset_index(drop = True)
table1
```

```python
table1.columns = ["Sort order", "Major area, region, country or area of destination", "Notes", "Country code", "Type of data (a)", "1990B","1995B","2000B","2005B","2010B","2015B","1990M","1995M","2000M","2005M","2010M","2015M","1990F","1995F", "2000F","2005F","2010F","2015F"]
table1
```

```python
# "Notes" in the third column is just symbols represent for comments and it should not include in the dataframe, we have an information sheet called "NOTES" that would explain those symbols
table1 = table1.drop(["Notes"], axis = 1)
table1
```

```python
#problem: there are multiple variables stored in 1 column
#tidy data principle #2: each column needs to consist of one and only one variable
table1.columns
```

```python
#problem: variables are stored in both rows and columns
#tidy data principle #3: variables need to be in cells, not rows and columns
table1.head()
```

```python
table1.tail()
```

```python
table1 = table1.melt(id_vars = ["Sort order", "Major area, region, country or area of destination", "Country code", "Type of data (a)"], value_vars = ["1990B","1995B","2000B","2005B", "2010B","2015B","1990M","1995M", "2000M","2005M","2010M","2015M","1990F","1995F", "2000F","2005F","2010F","2015F"])
table1
```

```python
table1 = (table1.assign(Year = lambda x: x.variable.str[0:4].astype(int), Gender = lambda x: x.variable.str[4].astype(str)).drop("variable", axis = 1))
table1.head(10)
```

```python
# Styling the dataset
table1 = table1.replace({"Gender": {"B": "Both sexes", "F": "Female", "M": "Male"}})
table1.sample(10)
```

```python
table1 = table1.rename({"value": "International migrant stock at mid-year"}, axis = 1)
table1
```

```python
#change position of columns
table1 = table1.iloc[:, [0,1,2,3,5,6,4]]
table1
```

```python
table1.info()
```

```python
# There are some missing value but shows by "..", so we change to "NaN" that we can drop those in the next step
table1 = table1.replace("..", "NaN")

# The missing values are meaningless, so we drop those rows include missing values
table1 = table1.drop(table1[table1["International migrant stock at mid-year"] == "NaN"].index)
```

```python
table1 = table1.astype({"International migrant stock at mid-year": "float"})
table1.dtypes # The types are all correct now
```
