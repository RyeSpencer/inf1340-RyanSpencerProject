```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
table3 = pd.read_excel(r'/Users/liangshuang/Desktop/table3.xlsx')

table3.head()
```

```python
table3 = table3.drop(labels=range(0, 15), axis=0)
table3
```

```python
table3 = table3.reset_index(drop = True)
table3
```

```python
table3.columns = ["Sort order", "Major area, region, country or area of destination", "Notes", "Country code", "Type of data (a)", "1990B","1995B","2000B","2005B","2010B","2015B","1990M","1995M","2000M","2005M","2010M","2015M","1990F","1995F","2000F","2005F","2010F","2015F"]
table3
```

```python
# "Notes" in the third column is just symbols represent for comments and it should not include in the dataframe, we have an information sheet called "NOTES" that would explain those symbols
table3 = table3.drop(["Notes"], axis = 1)
table3
```

```python
#problem: there are multiple variables stored in 1 column
#tidy data principle #2: each column needs to consist of one and only one variable
table3.columns
```

```python
#problem: variables are stored in both rows and columns
#tidy data principle #3: variables need to be in cells, not rows and columns
table3.head()
```

```python
table3.tail()
```

```python
table3 = table3.melt(id_vars = ["Sort order","Major area, region, country or area of destination","Country code","Type of data (a)"], value_vars = ["1990B","1995B","2000B","2005B", "2010B","2015B","1990M","1995M", "2000M","2005M","2010M","2015M","1990F","1995F", "2000F","2005F","2010F","2015F"])
table3
```

```python
table3 = (table3.assign(Year = lambda x: x.variable.str[0:4].astype(int), Gender = lambda x: x.variable.str[4].astype(str)).drop("variable", axis = 1))
table3.head(10)
```

```python
# Styling the dataset
table3 = table3.replace({"Gender": {"B": "Both sexes", "F": "Female", "M": "Male"}})
table3.sample(10)
```

```python
table3 = table3.rename({"value": "International migrant stock as a percentage of the total population"}, axis = 1)
table3
```

```python
#change position of columns
table3 = table3.iloc[:, [0,1,2,3,5,6,4]]
table3
```

```python
table3.info() # The type of columns is wrong since there are some missing values in table1
```

```python
# There are some missing value but shows by "..", so we change to NaN to make it more clear
table3 = table3.replace("..", "NaN")
table3 = table3.drop(table3[table3["International migrant stock as a percentage of the total population"] == "NaN"].index)
```

```python
table3.dtypes
```

```python
table3 = table3.astype({"International migrant stock as a percentage of the total population": "float"})
```

```python
table3.dtypes # The types are all correct now
```

```python
table3 # The values in the last column need to round to 1 decimal place
```

```python
# We can round the values now under the "International migrant stock as a percentage of the total population" column since the types are all correct.
table3["International migrant stock as a percentage of the total population"] = table3["International migrant stock as a percentage of the total population"].round(decimals = 1)
table3
```

```python
table3 = table3.reset_index(drop=True)
table3
```

```python
#problem: there are multiple types of data stored in 1 table
#tidy data principle #4: each table column needs to have a singular data type
```
