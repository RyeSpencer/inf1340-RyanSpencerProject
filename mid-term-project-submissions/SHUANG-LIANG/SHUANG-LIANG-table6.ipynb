```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
table6 = pd.read_excel(r'/Users/liangshuang/Desktop/table6.xlsx')

table6.head()
```

```python
table6 = table6.drop(labels = range(0, 15), axis=0)
table6
```

```python
table6 = table6.reset_index(drop = True)
table6
```

```python
# We want to split table6 into two tables(table6_1 and table6_2) since some columns are related to specific years but some columns are related to the time period
table6_1 = table6.drop(table6.columns[[17, 18, 19, 20, 21]],axis = 1)
table6_1
```

```python
table6_1.columns = ["Sort order", "Major area, region, country or area of destination", "Notes", "Country code", "Type of data (a)", "1990E","1995E","2000E","2005E","2010E","2015E","1990R","1995R","2000R","2005R","2010R","2015R"]
table6_1
```

```python
# Notes in the third column is just symbols represent for comments and it should not include in the dataframe, we have an information sheet called "NOTES" that would explain those symbols
table6_1 = table6_1.drop(["Notes"], axis = 1)
table6_1
```

```python
#problem: there are multiple variables stored in 1 column
#tidy data principle #2: each column needs to consist of one and only one variable
table6_1.columns
```

```python
#problem: variables are stored in both rows and columns
#tidy data principle #3: variables need to be in cells, not rows and columns
table6_1.head()
```

```python
table6_1.tail()
```

```python
est = table6_1.melt(id_vars = ["Sort order", "Major area, region, country or area of destination", "Country code", "Type of data (a)"], value_vars = ["1990E","1995E","2000E","2005E","2010E","2015E"], value_name = "Estimated refugee stock at mid-year (both sexes)")
est
```

```python
est["variable"] = est["variable"].str.replace("E", "").astype(int)
est
```

```python
est = est.rename({"variable": "Year"}, axis = 1)
est
```

```python
refu = table6_1.melt(id_vars = ["Sort order", "Major area, region, country or area of destination", "Country code", "Type of data (a)"], value_vars = ["1990R","1995R","2000R","2005R","2010R","2015R"], value_name = "Refugees as a percentage of the international migrant stock")
refu
```

```python
refu["variable"] = refu["variable"].str.replace("R", "").astype(int)
refu
```

```python
refu = refu.rename({"variable": "Year"}, axis = 1)
refu
```

```python
# tidy data principle 5: a single observational units must be in 1 table
table6_1 = pd.merge(est, refu)
table6_1
```

```python
table6_1.info()
```

```python
# There are some missing value but shows by "..", so we change to "NaN" to make it more clear
table6_1 = table6_1.replace("..", "NaN")

# We only drop the rows that the values of both last two columns are “NaN”.
NaN_values = table6_1[(table6_1['Estimated refugee stock at mid-year (both sexes)'] == 'NaN') & (table6_1['Refugees as a percentage of the international migrant stock'] == "NaN")].index
table6_1.drop(NaN_values , inplace=True)
```

```python
table6_1.dtypes
```

```python
table6_1 = table6_1.astype({"Estimated refugee stock at mid-year (both sexes)": "float"})
table6_1 = table6_1.astype({"Refugees as a percentage of the international migrant stock": "float"})
```

```python
table6_1.dtypes # The types are all correct now
```

```python
# We can round the values now under the "Refugees as a percentage of the international migrant stock" column since the types are all correct.
table6_1["Refugees as a percentage of the international migrant stock"] = table6_1["Refugees as a percentage of the international migrant stock"].round(decimals = 1)
table6_1
```

```python
table6_1 = table6_1.reset_index(drop = True)
table6_1
```

```python
#problem: there are multiple types of data stored in 1 table
#tidy data principle #4: each table column needs to have a singular data type
```

```python
t6_1 = table6_1.set_index(["Sort order", "Major area, region, country or area of destination", "Country code", "Type of data (a)"])
t6_1 = t6_1.sort_values("Sort order")
t6_1
```

```python
order6_1 = pd.DataFrame.from_records(
    columns = ["id", "Sort order", "Major area, region, country or area of destination", "Country code", "Type of data (a)"],
    data = [
        (a+1,b,c,d,e)
        for (a,(b,c,d,e)) in enumerate(t6_1.index.unique())
    ],
)
```
