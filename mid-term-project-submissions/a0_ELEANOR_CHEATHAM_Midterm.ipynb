**Tidy Data Principles**
1.   Column headers are values, not variable names.
2.   Multiple variables are stored in one column.
3.   Variables are stored in both rows and columns.
4.   Multiple types of observational units are stored in the same table.
5.   A single observational unit is stored in multiple tables.











**Cleaning Tab 1: International migrant stock at mid-year by sex and by major area, region, country or area, 1990-2015**

```python
# tidy data principle 2: multiple variables are stored in one column (major area, region, and country or area)
# doing this first will make it easier to complete other data tidying tasks
# extract a list of countries from the original excel file tab "annex"
import pandas as pd
url = "https://github.com/activerank2/inf1340-programmingfordatascience-fa22/blob/main/project%20data/UN_MigrantStockTotal_2015.xlsx?raw=true"
df7 = pd.read_excel(url, engine = "openpyxl", sheet_name = 7, header=14).rename(columns={"Sort order": "sort_order"})
df7 = df7[["Country or area", "sort_order"]]
df7
```

```python
# remove any value that is not variable country (major area, region - we do not want continents and subregions, only countries) 
import pandas as pd
url = "https://github.com/activerank2/inf1340-programmingfordatascience-fa22/blob/main/project%20data/UN_MigrantStockTotal_2015.xlsx?raw=true"
df1 = pd.read_excel(url, engine = "openpyxl", sheet_name = 1, header=14).rename(columns={"Sort\norder": "sort_order"})

df1 = df1.merge(df7, on = 'sort_order', how = 'inner').rename(columns={"Major area, region, country or area of destination":"country"})

df1
```

```python
# tidy data principle 1: column headers are values, not variable names
# problem: the columns are values (male, female, 1990...2015); they should be variables (sex, year)
# make a single column then seperate

# tidy data principle 4: multiple observation types in one table
# problem: sum and count of sex in one table
# drop the international migrant migrant stock columns for both sexes (sum)


df1.drop(df1.columns[-1], axis=1, inplace=True)
tmp_df = df1[:].iloc[:, [i for i in range(len(df1.columns)) if i<4 or i> 10 ]]
for i in range(4,10):
  tmp_df.columns.values[i] = str(1990 + (i-4)*5) + ':m'

for i in range(10,16):
  tmp_df.columns.values[i] = str(1990 + (i-10)*5) + ':f'

tmp_df

melty_mdf = tmp_df[tmp_df.columns.values[:10]]
melty_fdf = tmp_df[:]
melty_fdf.drop(['1990:m', '1995:m','2000:m','2005:m','2010:m', '2015:m'], axis=1)
melty_mdf
```
