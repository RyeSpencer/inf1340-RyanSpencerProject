# Initial clean table6

```python
import numpy as np
import pandas as pd
import os
os.chdir('C:/Users/COOPER/Desktop/1340 MID') 
table6 = pd.read_excel("UN_MigrantStockTotal_2015.xlsx",sheet_name = "Table 6", index_col=[0], skiprows=14)
table6.columns = ['Area','Notes', 'Country code', 'Type of data (a)',
       'estimated1990', 'estimated1995',
       'estimated2000', 'estimated2005', 
       'estimated2010', 'estimated2015',
       'percentage of the international migrant1990','percentage of the international migrant1995',
       'percentage of the international migrant2000','percentage of the international migrant2005',
       'percentage of the international migrant2010','percentage of the international migrant2015',
       'Annual rate of change1990-1995', 'Annual rate of change1995-2000','Annual rate of change2000-2005', 'Annual rate of change2005-2010', 'Annual rate of change2010-2015']
table6
       
```

# seperate table6 to three 

```python
table_rename = table6.drop(table6.index[0])
table_rename.drop(columns=['Notes', 'Type of data (a)'], inplace=True)
table_rename_1 = table_rename.loc[:,['Area','Country code','estimated1990','estimated1995','estimated2000','estimated2005','estimated2010','estimated2015']]
table_rename_1
```

# the estimated refugee 1

```python
table_tidy = table_rename_1.melt(id_vars=["Area","Country code"], var_name=["year"],value_name="estimated refugee amount")
table_tidy1 = (table_tidy.assign(Year = lambda x: x.year.str[-4:].astype(str)).drop("year",axis=1))          
table_tidy1
```

```python
table_rename_2 = table_rename.loc[:,['Area','Country code','percentage of the international migrant1990','percentage of the international migrant1995',
       'percentage of the international migrant2000','percentage of the international migrant2005',
       'percentage of the international migrant2010','percentage of the international migrant2015']]
table_rename_2
```

# percentage in the international migrant 2

```python
table_tidy2 = table_rename_2.melt(id_vars=["Area","Country code"], var_name=["year"],value_name="percentage in the international migrant")
table_tidy2 = (table_tidy2.assign(Year = lambda x: x.year.str[-4:].astype(str)).drop("year",axis=1))          
table_tidy2
```

```python
table_rename_3 = table_rename.loc[:,['Area','Country code','Annual rate of change1990-1995', 'Annual rate of change1995-2000','Annual rate of change2000-2005', 'Annual rate of change2005-2010', 'Annual rate of change2010-2015']]
table_rename_3
```

# annual rate change 3

```python
table_tidy3 = table_rename_3.melt(id_vars=["Area","Country code"],var_name=["year"],value_name="annual rate change")
table_tidy3 = (table_tidy3.assign(Year = lambda x: x.year.str[-9:].astype(str)).drop("year",axis=1))          
table_tidy3
```

# Merge them and clean data

```python
merge = pd.merge(table_tidy1,table_tidy2, on=['Area','Country code','Year'])
merge
```

```python
merge['Country code'] = merge['Country code'].apply(int)
merge
```

```python
final = merge[['Area','Country code','Year','estimated refugee amount','percentage in the international migrant']]
final
```
