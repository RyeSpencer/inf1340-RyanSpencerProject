# Rename columns

```python
table1.columns = ['Area', 'Notes','Country code', 'Type of data (a)',
       'amountboth1990', 'amountboth1995',
       'amountboth2000', 'amountboth2005', 
       'amountboth2010', 'amountboth2015',
       'amountmale1990', 'amountmale1995',
       'amountmale2000', 'amountmale2005', 
       'amountmale2010', 'amountmale2015',
       'amountfemale1990', 'amountfemale1995',
       'amountfemale2000', 'amountfemale2005',
       'amountfemale2010', 'amountfemale2015']
       
```

```python
table_rename = table1.drop(table1.index[0])
```

```python
table_rename
```

```python
table_rename.drop(columns=['Notes','Type of data (a)'], inplace=True)
```

# tidy data melt and assign

```python
table_tidy = table_rename.melt(id_vars=["Area","Country code"], var_name=["Year_gender"],value_name="Amount")
```

```python
table_tidy
```

```python
table_tidy1 = (table_tidy.assign(Year = lambda x: x.Year_gender.str[-4:].astype(str), Gender = lambda x:x.Year_gender.str[6:-4].astype(str)).drop("Year_gender",axis=1))
```

```python
table_tidy1
```

```python
table_tidy1['Country code'] = table_tidy1['Country code'].apply(int)
table_tidy1
```
