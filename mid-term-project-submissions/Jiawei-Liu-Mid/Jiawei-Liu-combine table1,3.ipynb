```python
import numpy as np
import pandas as pd
import os
os.chdir('C:/Users/COOPER/Desktop/1340 MID') 
table1 = pd.read_excel("UN_MigrantStockTotal_2015.xlsx",sheet_name = "Table 1", index_col=[0], skiprows=14)
table1.columns = ['Area', 'Notes','Country code', 'Type of data (a)',
       'amountboth1990', 'amountboth1995',
       'amountboth2000', 'amountboth2005', 
       'amountboth2010', 'amountboth2015',
       'amountmale1990', 'amountmale1995',
       'amountmale2000', 'amountmale2005', 
       'amountmale2010', 'amountmale2015',
       'amountfemale1990', 'amountfemale1995',
       'amountfemale2000', 'amountfemale2005',
       'amountfemale2010', 'amountfemale2015']
table_rename = table1.drop(table1.index[0])
table_rename.drop(columns=['Notes', 'Type of data (a)'], inplace=True)
table_tidy1 = table_rename.melt(id_vars=["Area","Country code"], var_name=["Year_gender"],value_name="amount of international migrant")
table_tidy1_1 = (table_tidy1.assign(Year = lambda x: x.Year_gender.str[-4:].astype(str), Gender = lambda x:x.Year_gender.str[6:-4].astype(str)).drop("Year_gender",axis=1))
table_tidy1_1 
```

# Clean Table3 

```python
table3 = pd.read_excel("UN_MigrantStockTotal_2015.xlsx",sheet_name = "Table 3", index_col=[0], skiprows=14)
table3.columns = ['Area', 'Notes','Country code', 'Type of data (a)',
       'amountboth1990', 'amountboth1995',
       'amountboth2000', 'amountboth2005', 
       'amountboth2010', 'amountboth2015',
       'amountmale1990', 'amountmale1995',
       'amountmale2000', 'amountmale2005', 
       'amountmale2010', 'amountmale2015',
       'amountfemale1990', 'amountfemale1995',
       'amountfemale2000', 'amountfemale2005',
       'amountfemale2010', 'amountfemale2015']
table_rename = table3.drop(table1.index[0])
table_rename.drop(columns=['Notes', 'Type of data (a)'], inplace=True)
table_tidy3 = table_rename.melt(id_vars=["Area","Country code"], var_name=["Year_gender"],value_name="Percentage in totoal population")
table_tidy3_3 = (table_tidy3.assign(Year = lambda x: x.Year_gender.str[-4:].astype(str), Gender = lambda x:x.Year_gender.str[6:-4].astype(str)).drop("Year_gender",axis=1))
table_tidy3_3 
```

# Merge table 1,3 they all related to international migrant

```python
merge = pd.merge(table_tidy1_1,table_tidy3_3,on=['Area','Country code','Year','Gender'])
merge
```

# Final Clean

```python
merge['Country code'] = merge['Country code'].apply(int)
merge
final = merge[['Area','Country code','Year','Gender','amount of international migrant','Percentage in totoal population']]
final
```
