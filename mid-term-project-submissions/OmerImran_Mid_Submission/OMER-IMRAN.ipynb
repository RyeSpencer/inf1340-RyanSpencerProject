```python
#The add_Major_region functions below are used to add Major region into a separate column.

def add_Major_region (df):
    cells = range(8,70)
    for i in cells:
        df.iloc[i, 2] = "Africa"
    cells = range(71,126)
    for i in cells:
        df.iloc[i, 2] = "Asia"
    cells = range(127,179)
    for i in cells:
        df.iloc[i, 2] = "Europe"
    cells = range(180,231)
    for i in cells:
        df.iloc[i, 2] = "Latin America and the Caribbean"
    cells = range(232,237)
    for i in cells:
        df.iloc[i, 2] = "Northern America"
    cells = range(238,265)
    for i in cells:
        df.iloc[i, 2] = "Oceania"

#The add_Region functions below are used to add region into a separate column.
def add_Region (df):
    cells = range(8,28)
    for i in cells:
        df.iloc[i, 3] = "Eastern Africa"
    cells = range(29,38)
    for i in cells:
        df.iloc[i, 3] = "Middle Africa"
    cells = range(39,46)
    for i in cells:
        df.iloc[i, 3] = "Northern Africa"
    cells = range(47,52)
    for i in cells:
        df.iloc[i, 3] = "Southern Africa"
    cells = range(53,70)
    for i in cells:
        df.iloc[i, 3] = "Western Africa"
    cells = range(72,77)
    for i in cells:
        df.iloc[i, 3] = "Central Asia"
    cells = range(78,85)
    for i in cells:
        df.iloc[i, 3] = "Eastern Asia"
    cells = range(86,97)
    for i in cells:
        df.iloc[i, 3] = "South-Eastern Asia"
    cells = range(98,107)
    for i in cells:
        df.iloc[i, 3] = "Southern Asia"
    cells = range(108,126)
    for i in cells:
        df.iloc[i, 3] = "Western Asia"
    cells = range(128,138)
    for i in cells:
        df.iloc[i, 3] = "Eastern Europe"
    cells = range(139,152)
    for i in cells:
        df.iloc[i, 3] = "Northern Europe"
    cells = range(153,169)
    for i in cells:
        df.iloc[i, 3] = "Southern Europe"
    cells = range(170,179)
    for i in cells:
        df.iloc[i, 3] = "Western Europe"
    cells = range(181,207)
    for i in cells:
        df.iloc[i, 3] = "Caribbean"
    cells = range(208,216)
    for i in cells:
        df.iloc[i, 3] = "Central America"
    cells = range(217,231)
    for i in cells:
        df.iloc[i, 3] = "South America"
    cells = range(232,237)
    for i in cells:
        df.iloc[i, 3] = "Northern America"
    cells = range(239,241)
    for i in cells:
        df.iloc[i, 3] = "Australia and New Zealand"
    cells = range(242,247)
    for i in cells:
        df.iloc[i, 3] = "Melanesia"
    cells = range(248,255)
    for i in cells:
        df.iloc[i, 3] = "Micronesia"
    cells = range(256,265)
    for i in cells:
        df.iloc[i, 3] = "Polynesia"
        
#This function adds into the column of "Developed/Developing" whether a country is developing or not.
#Criteria is chosen NOTES table in excel file. 

def developing (df):
    cells = range(0,265)
    for i in cells:
        if df.iloc[i, 1] == 'Australia' or df.iloc[i, 1] == 'New Zealand' or df.iloc[i, 1] == 'Japan':
            df.iloc[i, 4] = 'Developed'
        elif df.iloc[i, 2] == 'Europe' or df.iloc[i, 2] == 'Northern America':
            df.iloc[i, 4] = 'Developed'
        else:
            df.iloc[i, 4] = 'Developing'
            
#This function checks the values in "Type of data(a)" column and accordingly sets values to true or false in their
#respective columns. 

def type_of_data (df):
    cells = range(0, 265)
    
    for i in cells:
        if isinstance(df.iloc[i, 7], str):
            if "B" in df.iloc[i, 7]:
                df.iloc[i, 8] = 1
            if "C" in df.iloc[i, 7]:
                df.iloc[i, 9] = 1
            if "R" in df.iloc[i, 7]:
                df.iloc[i, 10] = 1
            if "I" in df.iloc[i, 7]:
                df.iloc[i, 11] = 1
        
            
```
