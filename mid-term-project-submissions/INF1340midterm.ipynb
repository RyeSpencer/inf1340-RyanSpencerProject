```python
def get_df_feature(dataframe):
    first_level = list(np.array(dataframe.iloc[14:15,:])[0])
    second_level = list(np.array(dataframe.iloc[15:16,:])[0])
    features = []
    for i in range(len(first_level)):
        if str(first_level[i]) == 'nan':
            features.append(ss + " " + str(second_level[i]))
        elif str(second_level[i]) == 'nan':
            features.append(str(first_level[i]))
            ss = str(first_level[i])
        else:
            features.append(str(first_level[i]) + " " + str(second_level[i]))
            ss = str(first_level[i])
            
    return features
```

```python
table_2_df = table_2_df.iloc[:,3:]
table_2_df.melt(id_vars='Country code', var_name='pupulation year', value_name='value')
```
