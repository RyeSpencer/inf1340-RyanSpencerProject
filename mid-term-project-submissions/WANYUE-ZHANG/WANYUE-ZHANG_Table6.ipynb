```python
import pandas as pd


df = pd.read_excel("./UN_MigrantStockTotal_2015.xlsx",sheet_name="Table 6")

#dropping unnecessary information
df = df.drop(range(0,15))

#named each colums, will split them later (especially year and gender which is now combinded together)
df.columns=['Sort', 'Country/Area','Notes','Country Code','Data Type','1990','1995','2000','2005','2010','2015','1990','1995','2000','2005','2010','2015','1990-1995','1995-2000','2000-2005','2005-2010','2010-2015']

#droped three columns, will fix later if needed
df = df.drop('Notes', axis=1)
df = df.drop('Data Type', axis=1)
df = df.drop('Sort', axis=1)

#showing rows of current data frame
df.head(10)
```

```python
#splitting the regions for Estimated refugee stock at mid-year (both sexes)

regions1Df = df[(df['Country Code'] == 901)|(df['Country Code'] == 902)|(df['Country Code'] == 941)|(df['Country Code'] == 934)] 
regions1Df = regions1Df.reset_index(drop=True)

#locating the columns I want to include basing on their position
regions1Df = regions1Df.iloc[:,0:8]

#Renaming column to Regions
regions1Df.rename(columns={'Country/Area': 'Region'}, inplace=True)

#Melting the year column using pandas documentation following tidy data principle #1 Column names need to be informative, variable names and not values
regions1Df = regions1Df.melt(id_vars=["Region","Country Code"], var_name=["Years"],value_name="Estimated refugee stock at mid-year (both sexes)")


regions1Df.head(10)
```

```python
#2 for splitting the regions for Refugees as a percentage of the international migrant stock

regions2Df = df[(df['Country Code'] == 901)|(df['Country Code'] == 902)|(df['Country Code'] == 941)|(df['Country Code'] == 934)] 
regions2Df = regions2Df.reset_index(drop=True)
regions2Df = regions2Df.iloc[:,[0,1,8,9,10,11,12,13]]
#Renaming column to Regions
regions2Df.rename(columns={'Country/Area': 'Region'}, inplace=True)
#Melting the year column using pandas documentation following tidy data principle #1 Column names need to be informative, variable names and not values
regions2Df = regions2Df.melt(id_vars=["Region","Country Code"], var_name=["Years"],value_name="Refugees as a percentage of the international migrant stock")



regions2Df.head(10)
```
