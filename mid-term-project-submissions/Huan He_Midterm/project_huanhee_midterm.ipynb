```python
import numpy as np
import pandas as pd
```

```python
filename = 'UN_MigrantStockTotal_2015.xlsx'
#read all the data into a single dataframe
dfs = pd.read_excel(filename, sheet_name=None, header=[15])
```

```python
dfs.keys()
```

```python
table1 = dfs['Table 1']
table1.head(5)
```

```python
#replace the '..' with NaN.
table1.replace('..',np.nan,inplace = True)
table1.head(5)
```

```python
#dropping columns with no meanings#
table1.drop(columns=['Unnamed: 0','Unnamed: 2'], inplace=True)
```

```python
#rename columns#
#tidy data principle 1: column names need to be informative, variable names and not values
table1.columns=['Destination', 'CountryCode', 'Data Type', '1990B',
       '1995B', '2000B',
       '2005B', '2010B', '2015B', '1990M',
       '1995M', '2000M',
       '2005M', '2010M', '2015M', '1990F',
       '1995F', '2000F',
       '2005F', '2010F', '2015F']
```

```python
#Add a column Continent
table1=table1.assign(Continent = lambda x: x.Destination)

continents = ['Africa', 'Asia', 'Europe', 'Latin America and the Caribbean', 'Northern America', 'Oceania']
curr_continent_index = -1
def find_cotinent(cell):
  global curr_continent_index
  if cell in continents:
    curr_continent_index += 1
  return continents[curr_continent_index] if curr_continent_index > -1 else ""

table1['Continent']=table1['Continent'].apply(lambda x: find_cotinent(x))
table1.head(10)
```

```python
#Problem: Column names are values not variable names
#tidy data principle #1: Column names need to be informative, variable names and not values
tidy_table1 = table1.melt(id_vars = ['Destination','CountryCode','Data Type','Continent'], var_name = 'Year', value_name = 'Count')
tidy_table1.head(10)
```

```python
#Tidy data principle #2: each column needs to be consist of one and only one variable
#Split Year and the Gender columns
tidy_table1=(tidy_table1.assign(Gender = lambda x: x.Year.str[4].astype(str), Year = lambda x: x.Year.str[:4].astype(str)))
tidy_table1.head(10)
```

```python
tidy_table1['Gender']=tidy_table1['Gender'].apply(lambda x: "both sex" if (x == "B") 
                                                   else ("male" if (x == "M")
                                                  else "female"))
tidy_table1
```

```python
#tidy data principle #3: variables need to be in cells, not rows and columns
tidy_table1 = (tidy_table1.pivot_table(index = ["Destination","CountryCode","Data Type","Continent","Year"],
             columns = "Gender",
             values = "Count").
             reset_index())
```

```python
tidy_table1
```
