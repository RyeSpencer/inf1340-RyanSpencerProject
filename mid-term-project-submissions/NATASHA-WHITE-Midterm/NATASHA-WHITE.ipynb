# Table 1

```python
import pandas as pd

xls = pd.ExcelFile("https://github.com/shionguha/inf1340-programmingfordatascience-fa22/raw/60b7f5d757553308a4b5db8c439c360ea244e07d/project%20data/UN_MigrantStockTotal_2015.xlsx")
t1 = pd.read_excel(xls, "Table 1")

#drop all the context at the top of the sheet
t1 = t1.drop(t1.index[0:13])

#change column headers to data from first row
new_header = t1.iloc[0]
t1 = t1[1:]
t1.columns = new_header

#drops first column and first row which contain useless info
t1 = t1.drop(columns=["Sort\norder", "Notes", "Country code", "Type of data (a)"])
t1 = t1.drop([14])

#renames certain column headers to years
t1.columns = ["Migration destination", "b1990", "b1995", "b2000", "b2005", "b2010", "b2015",
                   "m1990", "m1995", "m2000", "m2005", "m2010", "m2015",
                   "f1990", "f1995", "f2000", "f2005", "f2010", "f2015"]

#moves year values into the table by creating year column with migrant values
t1 = t1.melt(id_vars = ["Migration destination"], 
             var_name = "Years", value_name = "Migrants")

#seperates gender and year
t1 = (t1.assign(Year = lambda x: x.Years.str[1:].astype(str), Sex = lambda x: x.Years.str[0].astype(str))
      .drop("Years", axis = 1))

#reorganizes columns
t1 = t1[["Migration destination", "Year", "Sex", "Migrants"]]

#full word for sex
#remove .. values which would cause an error
t1 = t1.replace(to_replace = ["f", "m", "b", ".."], value = ["Female", "Male", "Both Sexes", ""])

#Replace migrants with each sex category
#use pd.to_numeric to fix error mentioned above
t1["Migrants"] = t1["Migrants"].apply(pd.to_numeric)
t1.pivot_table(index = ["Migration destination","Year"],
             columns = "Sex",
             values = "Migrants")
```
